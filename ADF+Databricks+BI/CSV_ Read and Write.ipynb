## **How to use multiple options**

### Options 
#### Header
#### InferSchema
#### Delimiter
#### nullValues
#### Dateformat
#### mode - PERMISSIVE, DROPMALFORMED & FAILFAST

```
options = {
    "inferSchema": "True",
    "delimiter": ",",
    "header": "True",
    "mode": "PERMISSIVE"
}
df_emp1 = spark.read.options(**options).csv("/FileStore/tables/department_data.txt")
```

```
df_emp1.display(10,truncate=False)
```

# **SAVE or WRITE: to CSV file(s)**

```
df = spark.read.format("csv").load("dbfs:/FileStore/tables/emp.txt")
df.show()
```

```
df_opt = spark.read.option('header', True) \
        .option('delimiter', ",") \
        .option('inferSchema', True) \
        .csv("dbfs:/FileStore/tables/emp.txt")
df_opt.display(truncate=False)
```

```
df_opt.write.options(header=True, delimiter="|") \
    .csv("dbfs:/FileStore/tables/output/emp.csv")
```

```
%fs ls "dbfs:/FileStore/tables/output"
```

```
df_opt.write.json("dbfs:/FileStore/tables/output/emp.json")
```

```
df_opt.write.parquet("dbfs:/FileStore/tables/output/emp.parquet")
```

```
df_opt.write.saveAsTable("EMP")
```

```
df_opt.write.mode("overwrite").saveAsTable("EMP")
```
